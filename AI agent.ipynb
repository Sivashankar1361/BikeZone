```python
class StudentAgent:
  def __init__(self,name):
    self.name=name
    self.previous_marks=None
    self.previous_attendance=None
    self.goal_attendance=80
    self.goal_marks=60

  def perceive(self,attendance,marks,study_hours):
    self.attendance=attendance
    self.marks=marks
    self.study_hours=study_hours

  def reflex_actions(self):
    actions=[]

    if self.attendance<75:
      actions.append("Increase classs attendance")

    if self.marks<40:
      actions.append("Revise fundamentals and practice more problems")

    return actions

  def model_based_reasoning(self):
    actions=[]

    if self.previous_marks is not None:
      if self.marks < self.previous_marks:
        actions.append("Marks are decreasing, follow a revision ")

    self.previous_marks=self.marks
    self.previous_attendance=self.attendance

    return actions

  def goal_based_reasoning(self):
    actions=[]

    if self.attendance<self.goal_attendance:
      actions.append("Attend extra classes to reach attendance goal")

    if self.marks < self.goal_marks:
      actions.append("Increase study hour to improve Marks")

    return actions

  def utility_based_decision(self,actions):
    best_action=None
    best_utility=-999

    for action in actions:
      utility=self.calculate_utility(action)
      if utility>best_utility:
        best_utility=utility
        best_action=action
    return best_action,best_utility

  def calculate_utility(self,action):
    if "attendace" in action:
      benifit=15
      effort=5
    elif "study" in action or "revise" in action:
      benifit=20
      effort=8
    else:
      benifit=10
      effort=4
    utility=benifit-effort
    return utility

  def act(self):
    print("\n--- Agent analysis Report ---")
    print(f"Student Name: {self.name}")
    print(f"Attendance: {self.attendance}%")
    print(f"Marks: {self.marks}")
    print(f"Study Hours/day: {self.study_hours}")

    reflex=self.reflex_actions()
    model =self.model_based_reasoning()
    goal=self.goal_based_reasoning()

    all_actions=list(set(reflex+model+goal))

    if not all_actions:
      print("\n Performance is satisfactory. keep it up!")
      return

    best_action,utility=self.utility_based_decision(all_actions)

    print("\n Possible Recommendations:")
    for act in all_actions:
      print(f"- {act}")

    print("\n Best Recommended Action")
    print(f"> {best_action}")
    print(f"Utility Score: {utility}")

def main():
    print("Intelligent Student Performance Advisor")
    print("---------------------------------------")

    name=input("Enter Student name:")
    attendance=float(input("Enter attendance percentage:"))
    marks=float(input("Enter current marks:"))
    study_hours=float(input("Enter study hours per day:"))

    agent=StudentAgent(name)
    agent.perceive(attendance,marks,study_hours)
    agent.act()


main()
```
